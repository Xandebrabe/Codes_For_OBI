# Lista encadeada

## Recapitulando o conceito de lista

Lista é uma sequência de elementos armazenados em posições contíguas de memória e que, portanto, ocupam um tamanho fixo. 

Listas são úteis nos mais variados cenários, mas há situações em que desejamos maior flexibilidade e as propriedades de arranjos não se adequam às nossas necessidades. 

## Um exemplo

Para um processo de negócio alcançar seu objetivo, define-se uma sequência de atividades que precisa ser executada. Por exemplo, para que a Anvisa aprove a comercialização de um novo produto de saúde, é necessário que uma série de atividades sejam executadas tanto pelo requisitante quanto pela Anvisa.

Nesse contexto, portanto, precisamos de uma estrutura de dados na qual uma atividade aponte para a atividade subsequente para que quando a primeira seja concluída, possamos saber qual a próxima tarefa a ser realizada.


Para casos como esse, precisamos de estruturas que nos permitam:

- inserir elementos no meio da estrutura de dados - se usarmos uma lista, a cada inserção desse tipo precisamos mover todos os elementos subsequentes ao elemento adicionado.

- inserir um número arbitrário de elementos - a estrutura de dados a ser utilizada precisa permitir tamanho dinâmico, sem valor máximo de quantidade de elementos pré-definido.

## Conceito de *lista encadeada*

Listas encadeadas são a forma mais simples de estrutura de dados dinâmica. Em uma lista encadeada, os elementos não são armazenados contiguamente, mas cada elemento é armazenado separadamente e possui valor (o dado que desejamos acessar) e um ponteiro para o próximo elemento. 

* Um ponteiro (ou apontador) é uma variável que armazena o endereço de outra variável (eles referenciam outra variável). No caso de listas encadeadas, um ponteiro armazena o endereço do próximo elemento da lista. 

Dessa forma, para percorrermos a lista, basta seguirmos os ponteiros ao longo dela.

> Desvantagens da lista encadeada
- espaço adicional para armazenar os ponteiros;
- para acessar um elemento em uma dada posição da lista, precisamos percorrer todos os elementos anteriores a ele na lista.

  - Essa cadeia de nós referenciando outro nó mostra o caráter encadeado da lista. 

- A classe Lista contém apenas um apontador para o primeiro elemento da lista. 

* O primeiro e o último elementos de uma lista são muito importantes, por isso recebem nomes especiais: cabeça e cauda da lista, respectivamente.
